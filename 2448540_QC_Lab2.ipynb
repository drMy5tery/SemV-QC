```python
# Quantum Computing Lab 2 - Basic Quantum Operations
# This lab explores fundamental single-qubit quantum gates and their properties

# Import necessary libraries
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Operator, Statevector
from qiskit_aer import AerSimulator
import numpy as np
import matplotlib.pyplot as plt

print("=" * 60)
print("Lab 2: Performing Operations with Operators and State Vectors")
print("=" * 60)

# ============================================================================
# Part 1: Execute operations on a state vector
# ============================================================================
print("\n--- Part 1: Executing Operations on State Vectors ---\n")

# Start with the |0⟩ state (ground state)
initial_state = Statevector.from_label('0')
print(f"Initial state |0⟩: {initial_state.data}")

# Let's apply each operation and see how it transforms the state

# X gate (Pauli-X / NOT gate) - flips |0⟩ to |1⟩ and vice versa
print("\n1. Applying X gate:")
x_operator = Operator.from_label('X')
state_after_x = initial_state.evolve(x_operator)
print(f"   State after X: {state_after_x.data}")
print(f"   This flips |0⟩ to |1⟩")

# Y gate (Pauli-Y) - rotation around Y-axis
print("\n2. Applying Y gate:")
y_operator = Operator.from_label('Y')
state_after_y = initial_state.evolve(y_operator)
print(f"   State after Y: {state_after_y.data}")
print(f"   Introduces both amplitude and phase changes")

# Z gate (Pauli-Z) - phase flip
print("\n3. Applying Z gate:")
z_operator = Operator.from_label('Z')
state_after_z = initial_state.evolve(z_operator)
print(f"   State after Z: {state_after_z.data}")
print(f"   Z doesn't change |0⟩, but would flip the phase of |1⟩")

# H gate (Hadamard) - creates superposition
print("\n4. Applying H gate:")
h_operator = Operator.from_label('H')
state_after_h = initial_state.evolve(h_operator)
print(f"   State after H: {state_after_h.data}")
print(f"   Creates equal superposition: (|0⟩ + |1⟩)/√2")

# S gate (Phase gate) - adds π/2 phase
print("\n5. Applying S gate:")
s_operator = Operator.from_label('S')
state_after_s = initial_state.evolve(s_operator)
print(f"   State after S: {state_after_s.data}")
print(f"   S adds a phase of π/2 to |1⟩ component")

# T gate (π/8 gate) - adds π/4 phase
print("\n6. Applying T gate:")
t_operator = Operator.from_label('T')
state_after_t = initial_state.evolve(t_operator)
print(f"   State after T: {state_after_t.data}")
print(f"   T adds a phase of π/4 to |1⟩ component")

# ============================================================================
# Part 2: Draw a simple circuit for these operations
# ============================================================================
print("\n\n--- Part 2: Drawing Quantum Circuits ---\n")

# Create a circuit with one qubit and one classical bit for measurement
qc = QuantumCircuit(1, 1)

# Apply all the gates in sequence
qc.x(0)  # Apply X gate
qc.y(0)  # Apply Y gate
qc.z(0)  # Apply Z gate
qc.h(0)  # Apply Hadamard gate
qc.s(0)  # Apply S gate
qc.t(0)  # Apply T gate

# Add measurement at the end
qc.measure(0, 0)

# Display the circuit
print("Circuit diagram with all operations:")
print(qc.draw(output='text'))

# ============================================================================
# Part 3: Simulate the experiment 10,000 times
# ============================================================================
print("\n\n--- Part 3: Running the Simulation (10,000 shots) ---\n")

# Initialize the simulator
simulator = AerSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the simulation with 10,000 shots
job = simulator.run(compiled_circuit, shots=10000)
result = job.result()

# Get the measurement counts
counts = result.get_counts(compiled_circuit)

print(f"Measurement results after 10,000 runs:")
print(f"  |0⟩ measured: {counts.get('0', 0)} times")
print(f"  |1⟩ measured: {counts.get('1', 0)} times")

# Calculate probabilities
total_shots = sum(counts.values())
prob_0 = counts.get('0', 0) / total_shots
prob_1 = counts.get('1', 0) / total_shots

print(f"\nProbabilities:")
print(f"  P(|0⟩) = {prob_0:.4f}")
print(f"  P(|1⟩) = {prob_1:.4f}")

# Visualize the results
plt.figure(figsize=(8, 6))
plt.bar(counts.keys(), counts.values(), color=['blue', 'red'])
plt.xlabel('Measurement Outcome', fontsize=12)
plt.ylabel('Number of Counts', fontsize=12)
plt.title('Quantum Circuit Measurement Results (10,000 shots)', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# Part 4: Proof that R (H-S-H) is the square root of NOT
# ============================================================================
print("\n\n--- Part 4: Proving R = √X (Square Root of NOT) ---\n")

print("We need to prove that R = H·S·H is the square root of X gate")
print("This means applying R twice should give us the X gate: R² = X\n")

# Define the R operation as H-S-H sequence
h_op = Operator.from_label('H')
s_op = Operator.from_label('S')

# Compose the operations: R = H · S · H
# Note: Operations are applied right to left in Qiskit's operator composition
R = h_op.compose(s_op).compose(h_op)

print("Step 1: Construct R = H·S·H")
print(f"R matrix:\n{R.data}\n")

# Now apply R twice: R²
R_squared = R.compose(R)

print("Step 2: Calculate R² (applying R twice)")
print(f"R² matrix:\n{R_squared.data}\n")

# Get the X gate matrix for comparison
x_gate = Operator.from_label('X')
print("Step 3: Compare with X gate")
print(f"X gate matrix:\n{x_gate.data}\n")

# Check if R² equals X (within numerical precision)
# We use np.allclose to handle floating point precision issues
if np.allclose(R_squared.data, x_gate.data):
    print("✓ PROOF SUCCESSFUL: R² = X")
    print("  This confirms that R = H·S·H is indeed the square root of NOT!")
else:
    print("✗ The matrices don't match exactly")
    difference = np.abs(R_squared.data - x_gate.data)
    print(f"  Maximum difference: {np.max(difference)}")

# Let's also verify this with state vectors
print("\nVerification with state vectors:")
test_state = Statevector.from_label('0')

# Apply R twice
state_after_R = test_state.evolve(R)
state_after_R_squared = state_after_R.evolve(R)

# Apply X once
state_after_X = test_state.evolve(x_gate)

print(f"  |0⟩ after R²: {state_after_R_squared.data}")
print(f"  |0⟩ after X:  {state_after_X.data}")

if np.allclose(state_after_R_squared.data, state_after_X.data):
    print("  ✓ State evolution confirms: R² = X")

# ============================================================================
# Additional visualization: R operation circuit
# ============================================================================
print("\n\nCircuit representation of R (square root of NOT):")
qc_r = QuantumCircuit(1)
qc_r.h(0)
qc_r.s(0)
qc_r.h(0)
print(qc_r.draw(output='text'))

print("\n" + "=" * 60)
print("Lab Assignment Complete!")
print("=" * 60)
```

```python
# Step 1: Importing the required libraries
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator, Statevector
from qiskit_aer import AerSimulator
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize a single qubit circuit
qc = QuantumCircuit(1)

# Apply the gates sequentially
qc.x(0)   # Pauli-X gate
qc.y(0)   # Pauli-Y gate
qc.z(0)   # Pauli-Z gate
qc.h(0)   # Hadamard gate
qc.s(0)   # S gate
qc.t(0)   # T gate

# Draw the circuit
qc.draw('mpl')
```

```python
# Get the final statevector
final_state = Statevector.from_instruction(qc)
print("Final Statevector:\n", final_state.data)
```

```python
# Add measurement
qc_measure = qc.copy()
qc_measure.measure_all()

# Use the qasm simulator
simulator = AerSimulator()

# Execute the circuit 10,000 times
job = simulator.run(qc_measure, shots=10000)
result = job.result()
counts = result.get_counts()

print("Measurement Counts:", counts)

# Visualize the result
plt.bar(counts.keys(), counts.values())
plt.xlabel('Measured State')
plt.ylabel('Counts')
plt.title('Measurement results after 10000 shots')
plt.show()
```

```python
# Define the gates
H = Operator.from_label('H')
S = Operator.from_label('S')
X = Operator.from_label('X')

# Construct R = H S H
# Note: In Qiskit, operator composition is applied right to left,
# but standard matrix multiplication (like @) applies left to right.
# To match the circuit order H-S-H applied to a state vector from the left,
# the matrix multiplication should be H @ S @ H.
R = H @ S @ H

# Check R^2
R_squared = R @ R

print("R (H S H) matrix:\n", np.round(R.data, 3))
print("\nR^2 matrix:\n", np.round(R_squared.data, 3))
print("\nX Gate matrix:\n", np.round(X.data, 3))

# Verify if R^2 ≈ X
is_square_root = np.allclose(R_squared.data, X.data)
print("\nIs R the square root of NOT?", is_square_root)
```
